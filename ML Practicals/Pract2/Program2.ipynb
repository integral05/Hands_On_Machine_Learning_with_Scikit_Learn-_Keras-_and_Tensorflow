### Spliting The Dataset 

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=54,test_size=0.30)
print("Train Size: ",x_train.shape)
print("Test Size: ",x_test.shape)
```

Using `KNeighborsClassifier` Classification

```python
from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(x_train,y_train)
```

```python
y_pred = kn.predict(x_test)
```

Checking Performance Of The KNN Classifier

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,confusion_matrix
```

```python
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
print("Accuracy: ",accuracy_score(y_test,y_pred))
```

```python
print("Precision: ",precision_score(y_test,y_pred))
```

```python
print("Recall Value: ",recall_score(y_test,y_pred))
```

```python
print("F1 Score: ",f1_score(y_test,y_pred))
```

Using `SVM` Model

```python
from sklearn.svm import SVC

svm = SVC(kernel='linear')
svm.fit(x_train,y_train)
```

```python
svm_y_pred = svm.predict(x_test)
```

Checking Performance Of The SVM Model

```python
confusion_matrix(y_test,svm_y_pred)
cm
```

```python
print("Accuracy: ",accuracy_score(y_test,svm_y_pred))
```

```python
print("Precision: ",precision_score(y_test,svm_y_pred))
```

```python
print("Recall Value: ",recall_score(y_test,svm_y_pred))
```

```python
print("F1 Score: ",f1_score(y_test,svm_y_pred))
```
